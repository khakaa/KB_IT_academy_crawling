```python
import requests
# 네이버 메인 뉴스목록 추출
url = 'https://www.naver.com/'

res = requests.get(url).text
res[:200]
```

```python
# 뉴스목록 파싱
from bs4 import BeautifulSoup

bs = BeautifulSoup(res, 'html.parser')

print(bs.title) # 항상 검색결과는 태그
print(type(bs.title))

# 글자 추출법
print(bs.title.string) # 1. 해당태그가 직접 가지고있는 글자
print(bs.title.get_text()) # 2. 해당태그내 모든 태그들이 가지고 있는 글자 
```

```python
news_list = bs.select('div.list_issue > div')
len(news_list)

news_title_arr = []

for news in news_list:
    news_title_arr.append(news.get_text())

news_title_arr
```

## KBO 일정 Ajax 추출

```python
url = 'https://www.koreabaseball.com/ws/Schedule.asmx/GetScheduleList'

# request data
params = {
    'leId': '1',
    'srIdList': '0,9',
    'seasonId' : '2022',
    'gameMonth': '07',
    'teamId': ''
}

sched_json = requests.post(url, params).json()
sched_json
```

```python
# print(len(sched_json['rows']))
# print(len(sched_json['rows'][0]['row']))
# print(len(sched_json['rows'][1]['row']))
# print(len(sched_json['rows'][2]['row']))
sched_dic_list = []

for s in sched_json['rows']:
    if s['row'][0]['Class'] == 'day':
        day = s['row'][0]['Text']
        
    sched_dic = {}
    
    for r in s['row']:
        # 일시
        sched_dic['일시'] = day

        # 시간
        if r['Class'] == 'time':
            bs = BeautifulSoup(r['Text'], 'html.parser')
            sched_dic['시간'] = bs.select_one('b').get_text()
        
        # 팀
        if r['Class'] == 'play':
            bs = BeautifulSoup(r['Text'], 'html.parser') 
            
            if len(bs.find_all('span')) == 5:
                sched_dic['팀'] = bs.find_all('span')[0].get_text() + ' : ' + bs.find_all('span')[4].get_text() 
                sched_dic['결과'] = bs.find_all('span')[1].get_text() + ' : ' + bs.find_all('span')[3].get_text()

    # 장소
    sched_dic['장소'] = s['row'][len(s['row']) - 2]['Text']

    if s['row'][len(s['row']) - 1]['Text'] == '-':
        sched_dic_list.append(sched_dic)

sched_dic_list
```
